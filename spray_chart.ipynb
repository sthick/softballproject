```python
import pandas as pd
```

```python
def fix_names(name):
    if name == 'Charla Echols':
        name = 'charla echols'
    elif name == 'Ebonee Echols':
        name = 'ebonee echols'
    elif name == 'Rylie Unzicker':
        name = 'rylie unzicker'
    elif name == 'madi unzicker':
        name == 'madi unzicker'
    elif name == 'Maeve Nelson':
        name = 'maeve nelson'
    else:
        name = name.split(' ')[1].lower()
    return name
    
def last_name(play):
    if 'C. Echols' in play:
        return 'charla echols'
    elif 'E. Echols' in play:
        return 'ebonee echols'
    elif 'M. Unzicker' in play:
        return 'madi unzicker'
    elif 'R. Unzicker' in play:
        return 'rylie unzicker'
    elif 'Ma. Nelson' in play:
        return 'maeve nelson'
    else:
        play_list = play.split(' ')
        for w in play_list:
            w = w.lower()
            if ',' in w:
                w = w.split(',')[0] 
            elif '-' in w:
                w = w.split('-')[0] 
            if w in last_name_list:
                return w

def fix_team(team):
    if team == 'Penn St.':
        team = 'Penn State'
    elif team == 'Ohio St.':
        team = 'Ohio State'
    elif team == 'Michigan St.':
        team = 'Michigan State'
    return team
```

```python
name_play = pd.read_csv('data/b10_dataframe.csv')
```

```python
name_play1 = name_play.copy()
name_play1['last_name'] = name_play1['play'].apply(last_name)
name_play1['team_name'] = name_play1['team_name'].apply(lambda x: x.strip())
name_play1['team_name'] = name_play1['team_name'].apply(fix_team)
```

```python
roster1 = pd.read_csv('https://raw.githubusercontent.com/sthick/softballproject/master/team_rosters_new.csv', header=None)
```

```python
roster = roster1.dropna()
roster
```

```python
roster['name'] = roster[0]
# name_series = pd.Series(['Thais Gonzalez', 'Michigan', 'Thais Gonzalez'], index = roster.columns)
# roster = roster.append(name_series, ignore_index=True)
roster['last_name'] = roster['name'].apply(fix_names)
# list of duplicate last names unzicker (same team), williams, edwards, echols (same team), bartz, nelson (same team)
roster = roster.rename(columns = {1:'team_name'})
```

```python
last_name_list = roster['last_name'].to_list()
```

```python
b10 = roster.merge(name_play1, how = 'left', on = ['last_name', 'team_name'])
```

```python
b10 = b10.drop(columns = [0, 'Unnamed: 0', 'name_y'])
```

```python
b10
```

```python
spraydf = b10.drop(columns = ['last_name', 'play', 'play_list', 'count_full', 
                              'count', 'pitches', 'pop_up', 'K', 'backwards_K', 'home_run',
                               'fly_ball', 'ground_out', 'lined_out', 'fouled_out', 'single',
                               'doubled', 'tripled', 'reached', 'walked', 'steal', 'scoring',
                               'quality_ab', 'end of inning'])
```

```python
spraydf = spraydf.drop(columns = ['Hits', 'Runs', 'Errors'])
```

```python
spraydf = spraydf.dropna(subset=['field_position'])
```

```python
def ur_x(x):
    if x == 1.0:
        return '(381,285)'
    else:
        return ''
def ul_x(x):
    if x == 1.0:
        return '(112,306)' 
    else:
        return ''
def uc_x(x):
    if x == 1.0:
        return '(256,349)'
    else:
        return ''
def lr_x(x):
    if x == 1.0:
        return '(327,182)'
    else:
        return ''
def ll_x(x):
    if x == 1.0:
        return '(181,174)'
    else:
        return ''
def lc_x(x):
    if x == 1.0:
        return '(253,146)'
    else:
        return ''
```

```python
spraydf
```

```python
spraydf['upper_right'] = spraydf['upper_right'].apply(ur_x)  
spraydf['upper_left'] = spraydf['upper_left'].apply(ul_x)
spraydf['upper_center'] = spraydf['upper_center'].apply(uc_x)
spraydf['lower_right'] = spraydf['lower_right'].apply(lr_x)
spraydf['lower_left'] = spraydf['lower_left'].apply(ll_x)
spraydf['lower_center'] = spraydf['lower_center'].apply(lc_x)
```

```python
spraydf['coord'] = spraydf['upper_right'] + spraydf['upper_left'] + spraydf['upper_center'] + spraydf['lower_right'] + spraydf['lower_left'] + spraydf['lower_center']
spraydf['x'] = spraydf['coord'].apply(lambda x: x.split(',')[0][1:])
spraydf['y'] = spraydf['coord'].apply(lambda x: x.split(',')[1][:-1])
```

```python
spraydf.columns
```

```python
spraydf_final = spraydf.drop(columns = ['upper_right', 'upper_left',
       'upper_center', 'lower_right', 'lower_left', 'lower_center', 'coord'])
```

```python
spraydf_final.to_csv('data/spraydf_final.csv')
```
